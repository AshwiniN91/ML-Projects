## Telecom Customer churn

Telecom customer churn is the phenomenon of customers switching from one telecom service provider to another.
It is important to understand telecom customer churn because it affects the revenue and profitability of telecom companies.

<blockquote>According to some sources, the average churn rate in the telecom industry is approximately `1.9% per month`,
but could rise as high as `67% annually` for prepaid services. Since the cost of acquiring new customers is up to
`25 times higher` than the cost of retaining them, fostering customer loyalty is key.</blockquote>

Some of the factors that influence telecom customer churn are
   <ul>
    <li>Customer  satisfaction</li>
    <li>Service quality</li>
    <li>Pricing</li>
    <li>Network coverage</li>
    <li>Customer service</li>
    <li>Loyalty programs</li>
    </ul>

Using advanced analytics, companies can identify at-risk customers and provide personalized and proactive customer service, reward customers for their tenure and referrals.

<div class="alert alert-block alert-info">
    <b>Objective: </b>
    To help telecom company to design and implement effective retention campaigns and interventions that can prevent or reduce customer churn and increase customer lifetime value using demographics, tenure, referrals, network usage information etc.
</div>

**Ashwini:  ANOVA Test: Used in comparing groups. If there are more than 2 groups then t-test is preferred over ANOVA**

```python
#Here is an ANOVA test to see if there is a significant difference between monthly charges and internet type

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd


model_cost = ols(formula = 'monthly_charge ~ internet_type', data=telecom_churn_df).fit()
anova_cost_results = sm.stats.anova_lm(model_cost, typ=2)

print("ANOVA Summary")
print(model_cost.summary())

print("ANOVA Results")
print(anova_cost_results)

#the large value for the F-test paired with a p-value >0.05 means that there is a significant difference between the monthly charges and internet types
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

final_pipe = ColumnTransformer([
   ('num', 'passthrough', filtered_numeric_columns),
   ('cat', ordinal_encoder, str_cols)
])


final_pipe.fit(features)
features_processed = final_pipe.transform(features)
```

```python
from sklearn.model_selection import train_test_split

features_train, features_test, target_train, target_test = train_test_split(features_processed, target, test_size = 0.3, random_state = 42)
```

## **Model Building**

```python
#First model is logistics regression

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(features_train, target_train)

predicted_label = model.predict(features_test)

```
